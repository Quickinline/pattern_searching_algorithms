```python
from pprint import pprint
def LevenshteinDistance( s:str, t:str):
  m = len(s)
  n = len(t)
  # // for all i and j, d[i,j] will hold the Levenshtein distance between
  # // the first i characters of s and the first j characters of t
  # declare int d[0..m, 0..n]

  d = [[] for i in range(m+1)]
  for i in range(m+1):
    d[i] = [0 for j in range(n+1)]
 
  for i in range(m+1):
    d[i][0] = i
  for i in range(n+1):
    d[0][i] = i

  outputSum = 0
  for i in range(1,m+1):
    for j in range(1,n+1):
      cost = 0 if s[i-1] == t[j-1] else 1
      d[i][j]= min([
                    d[i-1][j] + 1,
                    d[i][j-1] + 1,
                    d[i-1][j-1] + cost,
      ])
  pprint(d)
  return d[m][n]
```

# K medoid Implementation

```python
# import the dependencies
import pandas as pd
from pandas.core.frame import DataFrame as Df
```

```python
class KMedoids():
  # n_clusters: number of clusters
  # 
  def __init__(self, n_clusters, distance_function, data_cols, n_iterations = 5, initial_points = 'random'):
    # set the number of clusters to process
    self.n_clusters = n_clusters
    # specify the distance function between each two elements
    self.distance_function = distance_function
    # specify the list of column(s) that are relevant to the Kmedoids PAM
    self.cols = data_cols
    # set the initial centroids of the clusters or they're set as random
    self.initial_points = initial_points
    # set the number of iterations that the Kmedoid will repeat
    self.n_iterations = n_iterations

  
  @property
  def n_iterations(self):
    return self._n_iterations
  
  @n_iterations.setter
  def n_iterations (self, value:int):
    if value < 0:
      raise Exception("Number of iterations is negtive and needs to be positive")
    self._n_iterations = value

  @property
  def cols(self):
    return self._cols
  
  @cols.setter
  def cols(self, value):
    if type(value) is list:
      self._cols = value
    else:
      if type(value) is str:
        self._cols = [value]
      else:
        raise Exception("the specified column(s) is neither a list nor a str")
    self.dimension = len(value)

  @property
  def n_clusters(self):
    return self._n_clusters
  
  @n_clusters.setter
  def n_clusters(self, value):
    self._n_clusters = value
  
  @property
  def distance_function(self):
    return self._distance_function
  
  @distance_function.setter
  def distance_function(self, value):
    if  not callable(value):
      raise Exception("Distance function variable is not a function")
    if value.__code__.co_argcount != 2:
      raise Exception("Distance function doesn't have the right shape")
    self._distance_function = value
  
  @property
  def iterations(self):
    
    return self._iterations
  
  @iterations.setter
  def iterations(self, value:list):
    self._iterations = value
  
  def _assign_centroid(self,data:Df):
    
    def sanitize(row):
      return int(row['assigned_centroid'][len('dist_from_center_'):])
    
    output =  data.copy(True)
    # get the rows that contain the distances only
    distance_columns = data.filter(regex='^dist_from_center_')
    
    # assign each point to its closes centroid
    output['assigned_centroid'] = distance_columns.idxmin(axis = 1)

    # sanitize the centroid mention by removing the unnecessary prefix
    output['assigned_centroid'] = output.apply(sanitize,axis=1)
    
    return output
  
  def _compute_distances(self, data:Df, centroids:Df):
    output = data.copy(True)
    for index, centroid in centroids.iterrows():
      dist_colname = f'dist_from_center_{index}'
      output[dist_colname] = output.apply(lambda x: self.distance_function(centroid,x), axis=1)
    
    return output
  

  def _get_clusters(self, data:Df, centroids:Df):
    clusters= []
    for index,centroid in centroids.iterrows():
      clusters.append(data[data.assigned_centroid == index])
    
    return clusters
  
  def _compute_cost(self, clusters):
    sum = 0
    for clust in clusters:
      for index,row in clust.iterrows():
        sum+=row[f'dist_from_center_{row["assigned_centroid"]}']
    return sum
  
  # def _is_better(self, iterations):
  #   if not hasattr(self,'_first'):
  #     self._first = True
  #     return True
  #   else:
  #     return False
  
    

  def _select_centroid(self, data:Df, centroids):
    added_centroid = data.sample(1)
    centroids = centroids.sample(centroids.count()[0] - 1)
    
    return pd.merge(added_centroid, centroids, how= 'outer')
      
  def fit(self, data: Df):
    # we copy the data into another data frame that we process
    processing_df = data.copy(True)

    # we initialize the inital centroids randomly (or selected)
    if self.initial_points == 'random':
      # print("here")
      self.centroids = processing_df.sample(self.n_clusters)
    else:
      # TODO: implement manual centroid selection
      raise NotImplementedError()
    
    self.iterations = [{
        'centroids': self.centroids
    }]
    curr_itr = 0
    swap_cost = 1
    # while self._is_better(self.iterations[curr_itr]):
    while swap_cost > 0 :
      current_iteration = self.iterations[curr_itr]
      # processing_df = data.copy(True)
      # we select the non-centroid data from the dataset
      current_iteration['non_centroids'] = processing_df[~processing_df.index.isin(current_iteration['centroids'].index)]

      # we calculate the distance from the self.centroids to each point 
      current_iteration['non_centroids'] = self._compute_distances(current_iteration['non_centroids'], current_iteration['centroids'])
      
      # assign each point to a centroid
      current_iteration['non_centroids'] = self._assign_centroid(current_iteration['non_centroids'])
      
      # we get the individual clusters from the dataset
      current_iteration['clusters'] =  self._get_clusters(current_iteration['non_centroids'], current_iteration['centroids'])

      # we compute the cost
      current_iteration['cost'] = self._compute_cost(current_iteration['clusters'])
      
      new_centroids = self._select_centroid(current_iteration['non_centroids'], current_iteration['centroids'])

      # return new_centroids

      if curr_itr > 0:
        swap_cost = self.iterations[curr_itr]['cost'] - self.iterations[curr_itr - 1]['cost'] 

      if swap_cost > 0:
        curr_itr+= 1
        self.iterations.append({
            'centroids': new_centroids
        })
    return processing_df
```
